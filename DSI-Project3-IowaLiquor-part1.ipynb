```python
# Remove redundant columns

iowa.drop('County Number', axis=1, inplace=True)
iowa.head()
```

```python
iowa.drop('County', axis=1, inplace=True)
iowa.head()
```

```python
iowa.drop('Category', axis=1, inplace=True)
iowa.head()
```

```python
# Remove "$" prices from characters and convert values to floats.

iowa['State Bottle Cost'] = iowa['State Bottle Cost'].map(lambda x: x.lstrip('$'))
iowa['State Bottle Retail'] = iowa['State Bottle Retail'].map(lambda x: x.lstrip('$'))
iowa['Sale (Dollars)'] = iowa['Sale (Dollars)'].map(lambda x: x.lstrip('$'))
iowa.head()
```

```python
iowa[['State Bottle Cost', 'State Bottle Retail', 
      'Sale (Dollars)']] = iowa[['State Bottle Cost', 'State Bottle Retail', 'Sale (Dollars)']].astype(float)
iowa.head()
```

```python
# Convert dates
iowa['Date']=pd.to_datetime(iowa['Date'], infer_datetime_format=True)
iowa.head()
```

#### Compute New Columns and Tables

```python
# Margin and Price per liter
iowa['Margins'] = iowa['State Bottle Retail'] - iowa['State Bottle Cost']
iowa['Price/Liter'] = iowa['Sale (Dollars)'] / iowa['Volume Sold (Liters)']
iowa
```

##### We'll need to make a new dataframe that indexes quantities by store:
###### sales per store for all of 2015, sales per store for Q1 2015, sales per store for Q1 2016, total volumes sold, mean transaction revenue, gross margin, price per bottle, price per liter, etc., average sales per day, number of days open

##### Make sure to retain other variables that we'll want to use to build our models, such as zip code, county number, city, etc. We recommend that you spend some time thinking about the model you may want to fit and computing enough of the suggested quantities to give you a few options.

```python
# Sales per store, 2015

# Filter by our start and end dates
iowa.sort_values(by=["Store Number", "Date"], inplace=True)
start_date = pd.Timestamp("20150101")
end_date = pd.Timestamp("20151231")
mask = (iowa['Date'] >= start_date) & (iowa['Date'] <= end_date)
sales = iowa[mask]
sales.head()
```

```python
sales.to_csv('iowasales.csv')
```

```python
# Group by store name
sales = sales.groupby(by=["Store Number"], as_index=False)
# Compute sums, means
sales = sales.agg({"Sale (Dollars)": [np.sum, np.mean],
                   "Volume Sold (Liters)": [np.sum, np.mean],
                   "Margin": np.mean,
                   "Price per Liter": np.mean,
                   "Zip Code": lambda x: x.iloc[0], # just extract once, should be the same
                   "City": lambda x: x.iloc[0],
                   "County Number": lambda x: x.iloc[0]})
# Collapse the column indices
sales.columns = [' '.join(col).strip() for col in sales.columns.values]
# Rename columns

# Quick check
sales.head()
```
